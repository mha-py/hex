```python
import pyglet
from pyglet.gl import *
from pyglet.window import key
from pyglet.window import mouse

import numpy as np

%load_ext autoreload
%autoreload 2

from hex_show import *
import hex_show

_vis_consts = hex_show._vis_consts
```

```python
bsize = 8

# Initialization of the game
board = np.zeros((bsize, bsize), 'int')
turn = 1 # whos turn is it?
over = False


# Functions
def show_board(arr, numbers=False):
    n = len(arr)
    
    c = [(.2,.2,.2), (.6,.0,.0), (.0,.05,.85)]
    c = [np.array(cc) for cc in c]
    f = np.sqrt(3/4)
    
    
    ijmouse = pos_on_board(*mouse_pos)
    
    for i in range(n):
        for j in range(n):
            ij = np.asarray((i,j))
            xy = _vis_consts.ij2xy @ ij
            x, y = xy
            
            glBegin(GL_POLYGON)
            glColor3d(*c[arr[i,j]%3])
            if (i, j) == ijmouse:
                glColor3d(*(0.6*c[arr[i,j]%3]))
            
            for k in range(6):
                glVertex2i(int(50*(_vis_consts.hexagon[k,0]+x))+50, int(50*(_vis_consts.hexagon[k,1]+y)+50))
            glEnd()
            
            if numbers:
                label = pyglet.text.Label('%d, %d' %(i,j),
                                          font_name='Verdana',
                                          font_size=12,
                                          x=50*x+50, y=50*y+50,
                                          anchor_x='center', anchor_y='center')
                label.draw()
            


def winner_single(stones):
    '''Gives info if one player is connected left to right.
    stones: nxn-array of bool type, where the stones of the player lie.'''
    connected = np.zeros_like(stones) # welche Steine sind nach unten verbunden?
    con_new   = np.zeros_like(stones) # beim letzten update dazugekommen
    connected[0] = stones[0] # welche Steine sind nach unten verbunden?
    con_new[0] = stones[0] # Steine am unteren Rand sind automatisch verbunden.
    changed = True
    while changed:
        changed = False
        for i in range(N):
            for j in range(N):
                if con_new[i,j]:
                    # activated others around (i,j)
                    for (di, dj) in [(1,0), (0,1), (-1,1), (-1,0), (0,-1), (1,-1)]:
                        if 0<=i+di<N and 0<=j+dj<N:
                            if not connected[i+di,j+dj] and stones[i+di,j+dj]:
                                # nur wenn noch nicht als connected markieren weiter fortfahren
                                connected[i+di,j+dj] = True
                                con_new[i+di,j+dj] = True
                                changed = True
                    con_new[i,j] = False
    if np.any(connected[-1]):
        return True
    else:
        return False
    
    
def winner(board):
    if winner_single(board.T==1):
        return 1
    if winner_single(board==-1):
        return -1
    return 0
            
            

# Allows anti aliasing
config = pyglet.gl.Config(sample_buffers=1, samples=4)
window = pyglet.window.Window(config=config, resizable=False) 

#window = pyglet.window.Window()

@window.event
def on_mouse_press(x, y, button, modifiers):
    if button == mouse.LEFT:
        global board, turn, over
        #mouse_pos = np.asarray(((x-50)/50, (y-50)/50))
        #i, j = pos_on_board(mouse_pos)
        x, y = (x-50)/50, (y-50)/50
        i, j = pos_on_board(x, y)
        if over or i is None:
            return
        if board[i,j] != 0:
            return
        board[i,j] = turn
        turn *= -1
        if winner(board):
            over = True
        
mouse_pos = np.zeros(2)
        
@window.event
def on_mouse_motion(x, y, dx, dy):
    global mouse_pos
    mouse_pos = (x-50)/50, (y-50)/50
    #ijmouse = pos_on_board(mouse_pos)
    #print(ijmouse)
        
        
@window.event
def on_draw():

        # Clear buffers (necessary?)
        #glClear(GL_COLOR_BUFFER_BIT)
        #glEnable(GL_BLEND)                                                            
        #glEnable(GL_LINE_SMOOTH)
        #glHint (GL_LINE_SMOOTH_HINT, GL_DONT_CARE)
        glColor3d(0.8, 0.0, 0.0)

        show_board(board)

pyglet.app.run()
```
