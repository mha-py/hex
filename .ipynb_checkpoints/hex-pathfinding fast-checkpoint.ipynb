```python

def winner_single(stones):
    '''Gives info if one player is connected left to right.
    stones: nxn-array of bool type, where the stones of the player lie.'''
    connected = np.zeros_like(stones) # welche Steine sind nach unten verbunden?
    con_new   = np.zeros_like(stones) # beim letzten update dazugekommen
    connected[0] = stones[0] # welche Steine sind nach unten verbunden?
    con_new[0] = stones[0] # Steine am unteren Rand sind automatisch verbunden.
    changed = True
    while changed:
        changed = False
        for i in range(N):
            for j in range(N):
                if con_new[i,j]:
                    # activated others around (i,j)
                    for (di, dj) in [(1,0), (0,1), (-1,1), (-1,0), (0,-1), (1,-1)]:
                        if 0<=i+di<N and 0<=j+dj<N:
                            if not connected[i+di,j+dj] and stones[i+di,j+dj]:
                                # nur wenn noch nicht als connected markieren weiter fortfahren
                                connected[i+di,j+dj] = True
                                con_new[i+di,j+dj] = True
                                changed = True
                    con_new[i,j] = False
    if np.any(connected[-1]):
        return True
    else:
        return False
    
def winner__old(board):
    if winner_single(board.T==1):
        return 1
    if winner_single(board==-1):
        return -1
    return 0
```

```python
from numba import jit, njit

@njit(cache=True)
def dfs_fast(stones):
    '''The fast (jit) version of dfs for checking if bottom and top are connected. Here, the path cannot be returned.
    Time: ~5.4 µs'''
    Nx, Ny = stones.shape
    assert Nx==Ny, 'Your board is not quadratic?'
    N = Nx

    ## for dfs use a list; for bfs use a deque.
    frontier = [ (0, j) for j in range(N) if stones[0, j]==1 ]
    explored = []

    while len(frontier) > 0:
        (i, j) = frontier.pop()
        if i == N-1:
                return True
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 0<=i+di<N and 0<=j+dj<N:
                if stones[i+di, j+dj] == 1:
                    #if (i+di, j+dj) not in explored:
                    for expld in explored:
                        if expld == (i+di, j+dj):
                            break
                    else:
                        frontier.append((i+di, j+dj))
        explored.append((i, j))
    return False
            
def winner(board):
    if bdfs(board.T==1, path=False):
        return 1
    if bdfs(board==-1, path=False):
        return -1
    return 0

def winning_path(board):
    'Gives out the winning path for the one who ones'
    path = bdfs(board.T==1, path=True)
    if path:
        # Transpose it
        path = [ (j, i) for (i, j) in path ]
    if not path:
        path = bdfs(board==-1, path=True)
    return path
```

## Wert von Steinen ermitteln
... im Sinne von: Führt dieser Stein zum Sieg (Wert=1), führt er in Kombination mit einem zweiten Stein zum Sieg (Wert=1/4)?

```python
@njit
def value_of_stones(board):
    '''Gives a heuristic value of positions to set a stone on. Value of 1 means, that putting a stone there gives a win.
    Value of 0.25 means that a combination with another not yet set stone wins.
    The num2 states how many combinations with another stone give a winning combination of 2 new stones.
    '''
    inds_ij = [ (i, j) for i in range(bsize) for j in range(bsize) ]
    
    value1 = 1.*np.zeros_like(board)
    value2 = 1.*np.zeros_like(board)
    num2 = np.zeros_like(board)
    
    # Check if game is already over
    if not dfs_fast(board):
    
        # Check which stones make a win
        for i, j in inds_ij:
            if board[i,j]: continue
            board_mod = board.copy()
            board_mod[i,j] = 1
            if dfs_fast(board_mod>0):
                value1[i,j] = 1

        #for i, j in inds_ij:
        for k in range(len(inds_ij)):
            i, j = inds_ij[k]
            if board[i,j]: continue
            if value1[i,j]==1: continue
            #for ii, jj in inds_ij:
            for kk in range(k+1, len(inds_ij)):
                ii, jj = inds_ij[kk]
                if board[ii,jj]: continue
                if value1[ii,jj]==1: continue
                board_mod = board.copy()
                board_mod[i,j] = 1
                board_mod[ii,jj] = 1
                if dfs_fast(board_mod>0):
                    value2[i,j] = 1
                    value2[ii,jj] = 1
                    num2[i,j] += 1
                    num2[ii,jj] += 1

    return value1 + value2 / 4, num2
```

## Heuristik, um den Wert eines Boards anzugeben
... indem gemessen wird, wie viele Steine noch notwendig sind.<br>
Dazu wird die Entfernung zwischen beiden Enden (links und rechts) gemessen, wobei ein gelegter Stein die Distanz 0 hat, eine freie Zelle die Distanz 1, und ein Gegnerstein eine Distanz von unendlich hat. Dabei wird dieser Wert halbiert und den Kanten zwischen zwei benachbarten Zellen zu gerechnet.

```python
from numba import jit, njit

@njit
def dijkstra_dist(board):
    '''Measures the distance from one boarder to the other. Own stones have cost 0, grey stones have cost 1 and enemys stones have cost 1000.
    This can be used as a heuristic for the ai (you will probably use the inverted distance and take the difference of both players)
    Time: ~22µs'''
    N = len(board)
    b = np.ones((N+2, N))
    b[1:N+1] = board
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    v = (b==0) * 1 + (b==+1) * 0 + (b==-1) * 1000
    
    distances = np.Inf*np.ones((N+2, N))    # shortest distance to the node (i-1, j). (i==0 and i==N+1 are the border, 1<i<N+1 are the cells)
    distances[0, :] = 0.                    # distance is 0 for starting points
    inqueue = np.zeros((N+2, N))
    inqueue[0, :] = True
    
    while np.any(inqueue):
        notinqueue = 1-inqueue
        k = np.argmin(distances + np.Inf*notinqueue)
        i, j = k//N, k%N
        
        dist = distances[i,j]
        inqueue[i,j] = False
        
        #if i == N+1:
        #    return dist
        
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 1<=i+di<N+2 and 0<=j+dj<N:             # streng genommen wäre die untere Grenze 0, aber dort ist aber der Start
                d = dist + v[i,j]/2 + v[i+di,j+dj]/2  # Anwärter für die Distanz zu (i+di,j+dj)
                if d < distances[i+di, j+dj]:
                    distances[i+di, j+dj] = d         # Wert überschreiben, da Distanz nun geringer
                    inqueue[i+di, j+dj] = True        # Knoten in die Queue hinzufügen
                    
    return np.min(distances[N+1,:])
       
    


@njit
def dijkstra_dist(board):
    '4 µs für 6x6'
    N = len(board)
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    MAXCOST = 100000
    v = (board==0) * 1 + (board==+1) * 0 + (board==-1) * 2*MAXCOST
    
    distances = MAXCOST*np.ones((N, N))    # shortest distance to the node (i-1, j). (i==0 and i==N+1 are the border, 1<i<N+1 are the cells)
    distances[0, :] = v[0,:]/2            # distance is the step from the edge to the first stone for starting points
    
    queue = [ (v[0,j]/2, (0,j)) for j in range(N) ]
    heapify(queue)
    
    dconnect = np.Inf
    
    while len(queue)>0:
        dist, (i, j) = heappop(queue)
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 1<=i+di<N and 0<=j+dj<N:             # streng genommen wäre die untere Grenze 0, aber dort ist aber der Start
                d = dist + v[i,j]/2 + v[i+di,j+dj]/2  # Anwärter für die Distanz zu (i+di,j+dj)
                if d < distances[i+di, j+dj]:
                    distances[i+di, j+dj] = d         # Wert überschreiben, da Distanz nun geringer
                    heappush(queue, (d, (i+di,j+dj))) # Knoten in die Queue hinzufügen
            elif i+di==N:
                if d + v[i,j]/2 < dconnect:
                    dconnect = d + v[i,j]/2
                    break
                else:
                    return dconnect
    
    
    
@njit
def heuristic_value(board):
    '''Gives a heuristic estimate value for the value of the board.
    A value near +1 means red is going to win, value near -1 means
    blue is going to win.
    This implementation basically counts the number of stones missing
    for a win and compares both players numbers.
    This algorithm can also return if a player has actually won, but
    dfa_fast is much faster than this.'''
    dist_red = dijkstra_dist(board.T)
    dist_blue = dijkstra_dist(-board)
    ##print(dist_red, dist_blue)
    if dist_red==0:  return +1
    if dist_blue==0: return -1
    return 1./(dist_red) - 1./(dist_blue)
```

```python


from numba import int8

def dijkstra_dist_with_path(board, freecells=True):
    '''Measures the distance from one boarder to the other.
    Additionally to the basic version of dijkstra, it returns all cells which are a part of one of the shortes paths.
    freecells: Only the cells which have to be taken are highlighted.
    Time: ~40µs
    '''
    N = len(board)
    b = np.ones((N+2, N))
    b[1:N+1] = board
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    v = (b==0) * 1 + (b==+1) * 0 + (b==-1) * 1000
    
    distances = np.Inf*np.ones((N+2, N))    # shortest distance to the node (i-1, j). (i==0 and i==N+1 are the border, 1<i<N+1 are the cells)
    distances[0, :] = 0.                    # distance is 0 for starting points
    inqueue = np.zeros((N+2, N))
    inqueue[0, :] = True
    ##reachedby = [ [ [ (0,0) for k in range(0) ] for j in range(N) ] for i in range(N+2) ]  # the preceders of (i,j) will be saves in 
    reachedby = -np.ones((N+2, N, 6, 2), dtype=int8)
    
    while np.any(inqueue):
        k = np.argmin(distances + 1e5*(1-inqueue))
        
        i, j = k//N, k%N
        
        dist = distances[i,j]
        inqueue[i,j] = False
        
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 1<=i+di<N+2 and 0<=j+dj<N:              # streng genommen wäre die untere Grenze für i die 0, aber dort ist aber der Start
                d = dist + v[i,j]/2 + v[i+di,j+dj]/2   # Anwärter für die Distanz zu (i+di,j+dj)
                if d < distances[i+di, j+dj]:
                    distances[i+di, j+dj] = d          # Wert überschreiben, da Distanz nun geringer
                    inqueue[i+di, j+dj] = True         # Knoten in die Queue hinzufügen
                    reachedby[i+di,j+dj,0] = (i,j)     # Dieser Knoten wurde (auf schnellstem Weg) nur von (i,j) aus erreicht.
                    reachedby[i+di,j+dj,1:] = -1       # ansonsten die Liste 'leeren'
                elif d == distances[i+di, j+dj]:
                    for l in range(6):                 # (ersten freien Platz der Liste finden)
                        if reachedby[i+di,j+dj,l,0] == -1:
                            break
                    reachedby[i+di,j+dj,l] = (i,j)    # Dieser Knoten wird auch durch (i,j) schnellstens erreicht.
                   
    distance = np.min(distances[N+1,:])
                    
    # Find  all cells which are on one of the shortest paths.
    # Uses dfs.
    on_path = np.zeros((N+2, N))
    frontier = [ (N+1, j) for j in range(N) ]
    visited = np.zeros((N+2, N))
    while len(frontier)>0:
        (i, j) = frontier.pop()
        if visited[i,j]:
            continue
        on_path[i,j] = True
        visited[i,j] = True
        parent_nodes = reachedby[i][j]
        for ii, jj in reachedby[i,j]:
            if ii < 0:
                break
            if visited[ii,jj]:
                continue
            frontier += [ (ii, jj) ]
        ##frontier.extend( [ (ii, jj) for (ii, jj) in parent_nodes if not visited[ii,jj] ] )
    
    # Cut the imaginary stones on the border off
    on_path = on_path[1:N+1]
    
    # If wanted, select the cells which are free
    if freecells:
        on_path *= (board == 0)
    
    return distance, on_path
                    
```

```python


from numba import int8

@njit
def dijkstra_dist_with_path_changes(oldboard, ij, newcell, distances, reachedby, freecells=True):
    # ist nie fertig gestellt worden..
    '''Measures the distance from one boarder to the other.
    Additionally to the basic version of dijkstra, it returns all cells which are a part of one of the shortes paths.
    freecells: Only the cells which have to be taken are highlighted.
    Time: ~40µs
    '''
    i0, j0 = ij
    oldcell = oldboard[i0,j0]
    board = oldboard.copy()
    board[i0,j0] = newcell
    
    if oldcell == 0:
        oldcost = 1
    elif oldcell == 1:
        oldcost = 0
    elif oldcell == -1:
        oldcost = 1000
        
    if newcell == 0:
        newcost = 1
    elif newcell == 1:
        newcost = 0
    elif newcell == -1:
        newcost = 1000
        
    
    N = len(board)
    b = np.ones((N+2, N))
    b[1:N+1] = board
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    v = (b==0) * 1 + (b==+1) * 0 + (b==-1) * 1000
    
    inqueue = np.zeros((N+2, N))
    
    distances = distances.copy()
    distances[i0+1, j0] += (newcost - oldcost)/2
    
    if newcost < oldcost:
        
        inqueue[i0+1, j0] = True
        
        while np.any(inqueue):
            # wie vorher auch alle veränderten updaten (wie bei dijkstra normal)
            notinqueue = 1-inqueue
            k = np.argmin(distances + 10000000*notinqueue)
            i, j = k//N, k%N

            dist = distances[i,j]
            inqueue[i,j] = False
            
            ####print('working on ', i, j)

            for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
                if 1<=i+di<N+2 and 0<=j+dj<N:
                    
                    d = dist + v[i,j]/2 + v[i+di,j+dj]/2   # Anwärter für die Distanz zu (i+di,j+dj)
                    #print((i+di,j+dj))
                    #print(d, distances[i+di, j+dj])
                    if d < distances[i+di, j+dj]:
                        distances[i+di, j+dj] = d          # Wert überschreiben, da Distanz nun geringer
                        inqueue[i+di, j+dj] = True         # Knoten in die Queue hinzufügen
                        reachedby[i+di,j+dj,0] = (i,j)     # Dieser Knoten wurde (auf schnellstem Weg) nur von (i,j) aus erreicht.
                        reachedby[i+di,j+dj,1:] = -1       # ansonsten die Liste 'leeren'
                    elif d == distances[i+di, j+dj]:
                        for l in range(6):                 # (ersten freien Platz der Liste finden)
                            if reachedby[i+di,j+dj,l,0] == -1:
                                break
                        reachedby[i+di,j+dj,l] = (i,j)     # Dieser Knoten wird auch durch (i,j) schnellstens erreicht.

    else: # if newcost < oldcost
        
        inqueue[i0, j0] = True
        
        while np.any(inqueue):
            notinqueue = 1-inqueue
            k = np.argmin(distances + 10000000*notinqueue)
            i, j = k//N, k%N
            
            inqueue[i,j] = False
            
            print('working on', (i,j))
            
            # erster Schritt: Die eigene Distanz und reachedby bestimmen (außer bei i0,j0 (dort Änderung trivial und oben schon erledigt)).
            # dieser Teil des Codes ist dem der Hauptfunktion ähnlich, vertauscht aber i,j und i+di,j+dj!
            if (i,j) != (i0,j0):
                for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
                    if 1<=i+di<N+2 and 0<=j+dj<N:
                        d = distances[i+di, j+dj] + v[i,j]/2 + v[i+di,j+dj]/2   # Anwärter für die Distanz
                        if d < distances[i, j]:
                            distances[i, j] = d                # Wert überschreiben, da Distanz nun geringer
                            reachedby[i,j,0] = (i+di,j+dj)     # Dieser Knoten wurde (auf schnellstem Weg) nur von (i,j) aus erreicht.
                            reachedby[i,j,1:] = -1             # ansonsten die Liste 'leeren'
                        elif d == distances[i, j]:
                            for l in range(6):                 # (ersten freien Platz der Liste finden)
                                if reachedby[i,j,l,0] == -1:
                                    break
                            reachedby[i,j,l] = (i+di,j+dj)     # Dieser Knoten wird auch durch (i,j) schnellstens erreicht.
                        
            # zweiter Schritt: Die Nachbarknoten in die Queue packen, falls sie sich auf i,j bezogen haben.
            for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
                if 1<=i+di<N+2 and 0<=j+dj<N:
                    for l in range(6):
                        if reachedby[i+di,j+dj,l,0] == -1:
                            break
                        if np.all(reachedby[i+di,j+dj, l] == (i, j)):
                            if l == 0:
                                # i, j was the only one who minimized the distance of i+di,j+dj
                                # so put him into the queue to fix his values
                                reachedby[i+di,j+dj,l] = (-1, -1)
                                distances[i+di,j+dj] = np.Inf
                                inqueue[i+di,j+dj] = True
                            else:  # if l == 0
                                reachedby[i+di,j+dj, l:-1] = reachedby[i+di,j+dj, l+1:]
                                reachedby[i+di,j+dj, -1] = (-1, -1)
                            
    distance = np.min(distances[N+1,:])
    
    return distance, _, distances, reachedby
                    
```

# Elektrische Spannung zwischen den Enden

```python
from numba import jit, njit
# SICHERUNG VOR OPTIMIERUNG 4.4.2021
@njit
def dijkstra_map(board):
    '180 µs für 13x13'
    N = len(board)
    b = np.zeros((N+1, N)) # zeros!
    b[0,:] = 1.
    b[1:N+1] = board
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    v = (b==0) * 1 + (b==+1) * 0 + (b==-1) * 1000
    
    distances = np.Inf*np.ones((N+1, N))    # shortest distance to the node (i-1, j). (i==0 and i==N+1 are the border, 1<i<N+1 are the cells)
    distances[0, :] = 0.                    # distance is 0 for starting points
    inqueue = np.zeros((N+1, N))
    inqueue[0, :] = True
    
    while np.any(inqueue):
        notinqueue = 1-inqueue
        k = np.argmin(distances + np.Inf*notinqueue)
        i, j = k//N, k%N
        
        dist = distances[i,j]
        inqueue[i,j] = False
        
        #if i == N+1:
        #    return dist
        
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 1<=i+di<N+1 and 0<=j+dj<N:             # streng genommen wäre die untere Grenze 0, aber dort ist aber der Start
                d = dist + v[i,j]/2 + v[i+di,j+dj]/2  # Anwärter für die Distanz zu (i+di,j+dj)
                if d < distances[i+di, j+dj]:
                    distances[i+di, j+dj] = d         # Wert überschreiben, da Distanz nun geringer
                    inqueue[i+di, j+dj] = True        # Knoten in die Queue hinzufügen
             
    return distances[1:] # Erweiterung an den Seiten wieder wegnehmen
                    
def voltage(board):
    A = dijkstra_map(board)
    B = dijkstra_map(board[::-1,::-1])[::-1, ::-1]
    return A/(A+B+1e-3)*2-1
```

```python
from numba import jit, njit
import numba
from heapq import heappush, heappop, heapify

                    

@njit
def dijkstra_map(board):
    '25 µs für 13x13'
    N = len(board)
    
    # Gegnerstein ist unpassierbar: kostet 1000
    # Leere Zelle muss noch besetzt werden: kostet 1
    # eingenommene Zelle: kostet 0
    MAXCOST = 100000
    v = (board==0) * 1 + (board==+1) * 0 + (board==-1) * 2*MAXCOST
    
    distances = MAXCOST*np.ones((N, N))    # shortest distance to the node (i-1, j). (i==0 and i==N+1 are the border, 1<i<N+1 are the cells)
    distances[0, :] = v[0,:]/2            # distance is the step from the edge to the first stone for starting points
    
    queue = [ (v[0,j]/2, (0,j)) for j in range(N) ]
    heapify(queue)
    
    while len(queue)>0:
        dist, (i, j) = heappop(queue)
        for (di, dj) in [(1,0), (1,-1), (0,1), (0,-1), (-1,1), (-1,0)]:
            if 1<=i+di<N and 0<=j+dj<N:             # streng genommen wäre die untere Grenze 0, aber dort ist aber der Start
                d = dist + v[i,j]/2 + v[i+di,j+dj]/2  # Anwärter für die Distanz zu (i+di,j+dj)
                if d < distances[i+di, j+dj] and d < MAXCOST:
                    distances[i+di, j+dj] = d         # Wert überschreiben, da Distanz nun geringer
                    heappush(queue, (d, (i+di,j+dj))) # Knoten in die Queue hinzufügen
             
    return distances # Erweiterung an den Seiten wieder wegnehmen
                    
def voltage(board):
    A = dijkstra_map(board)
    B = dijkstra_map(board[::-1,::-1])[::-1, ::-1]
    return A/(A+B+1e-3)*2-1
```
